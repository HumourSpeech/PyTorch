<h4> Tensors in PyTorch</h4>

Tensors are the core data structure in PyTorch.
They are similar to NumPy arrays but support:
- GPU acceleration
- Automatic differentiation

Example:

```python
import torch

x = torch.tensor([[1, 2], [3, 4]])
print(x)
print(x.shape)
print(x.size())
```

```python
# From Python list
a = torch.tensor([1, 2, 3])

# Zeros and ones
b = torch.zeros(3, 4)
c = torch.ones(2, 2)

# Random values
d = torch.rand(3, 3)
e = torch.randn(3, 3)
# randn() - normal distribution
# rand() - uniform distribution

# similar to NumPy
f = torch.randn(5)
```

```python
a, b, c, d, e, f
```
